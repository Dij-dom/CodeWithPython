
Missing Data (VERY COMMON)

Find missing values in a DataFrame.



```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [25, np.nan, 35, np.nan],
    "salary": [50000, "NA", 70000, 60000]
})

df.head()
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```


Fill missing values with mean/median.

```python
df.head(5)
```

Replace specific values (e.g., "NA" â†’ np.nan).

```python
df.replace("NA", np.nan, inplace=True)
df.head()
```

```python
df.isna().sum()
```

Drop rows with missing values.

| Situation                   | Use    |
| --------------------------- | ------ |
| Numeric + symmetric data    | Mean   |
| Numeric + skewed / outliers | Median |
| Categorical                 | Mode   |


```python
df["age"] = df["age"].fillna(df["age"].median())
df.head()
```
