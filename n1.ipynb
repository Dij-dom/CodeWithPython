Reverse a list without using .reverse() or slicing.


```python
# method 1
l = [x for x in range(0,10)]
print(f"Original List: {l}")
p1, p2 = 0, len(l)-1
while p1<p2:
    l[p1], l[p2] = l[p2], l[p1]
    p1 +=1
    p2 -=1
print(f"Recersed List: {l}")
```


Remove duplicates from a list while preserving order.


```python
lst = [1,1,2]
seen = set()
result_list = []

for i in lst:
    if i not in seen:
        seen.add(i)
        result_list.append(i)
# result_list = list(seen)
result_list
```
