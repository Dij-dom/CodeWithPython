Find all pairs with sum = K.

```python
nums = [1, 4, 2, 3, 5]
K = 5
pairs = []

unique_nums = set()
for num in nums:
    target = K - num
    if target in unique_nums:
        pairs.append((target, num))
    unique_nums.add(num)

pairs
```

Move all zeros to the end of an array.




```python
lst = [1, 0, 12, 0, 10]
count  = 0
result = []
for num in lst:
    if num == 0:
        count +=1
    else:
        result.append(num)
result.extend([0]*count)
result
```

Check if array is sorted.



```python
arr = [1, 3, 2]
flag =  True

for i in range(len(arr) - 1):
    if arr[i] > arr[i + 1]:
        flag = False

flag
```

    Find duplicate elements.

```python
arr = [1, 3, 2, 2]
seen = set()
duplicates = set()

for x in arr:
    if x in seen:
        duplicates.add(x)
    else:
        seen.add(x)
duplicates
```
