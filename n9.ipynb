GroupBy & Aggregations (CRITICAL)

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "dept": ["HR", "HR", "IT", "IT", "IT", "HR"],
    "gender": ["F", "M", "M", "M", "F", "M"],
    "salary": [50000, 60000, 70000, 80000, 75000, 65000]
})

df
```

Group by a column and calculate mean.

```python
df.groupby("dept")["salary"].mean()
```

Group by multiple columns.

```python
df.groupby(["dept", "gender"])["salary"].mean()
```

Find count per category.

```python
df.groupby("dept")["name"].count()
```

Find top N values per group.

```python
df.sort_values("salary", ascending=False).groupby("dept").head(2)
```

Aggregate using multiple functions.


```python
df.groupby("dept")["salary"].agg(["mean", "max"])
```
