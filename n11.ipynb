Merge & Join

```python
import pandas as pd

employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "David"],
    "dept_id": [10, 20, 10, 30]
})

departments = pd.DataFrame({
    "dept_id": [10, 20, 40],
    "dept_name": ["HR", "IT", "Finance"]
})

employees.head()
```

```python
departments.head()
```

Inner join two DataFrames.




```python
pd.merge(employees, departments, on="dept_id", how="inner")
```

Left join vs inner join â€” difference.



```python
pd.merge(employees, departments, on="dept_id", how="left")
```


Remove duplicates.


```python
employees.drop_duplicates(subset=["emp_id"])
```


Concatenate DataFrames.

```python
df1 = pd.DataFrame({
    "emp_id": [1, 2],
    "name": ["Alice", "Bob"]
})

df2 = pd.DataFrame({
    "emp_id": [3, 4],
    "name": ["Charlie", "David"]
})
```

```python
df1.head()
```

```python
df2.head()
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1, ignore_index=True)
```
