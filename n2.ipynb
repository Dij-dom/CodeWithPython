Check if a string is a palindrome (ignore spaces & case).

```python
s = "Race Car"
cleaned = ""
for ch in s:
    print(ch)
    if ch != " ": # ignores space
        cleaned += ch.lower() # converts to lowercase
cleaned[::-1] == cleaned
```

Find the first non-repeating character in a string.

```python
count_dict = {}

s = "aabbcddeefggh"

for ch in s:
    count_dict[ch] = count_dict.get(ch, 0) + 1

for ch in s:
    if count_dict[ch] ==1:
        print(f"The first non-repeating character is {ch}")
        break
```


Check if two strings are anagrams.




```python
from collections import Counter
s1 = "Listen"
s2 = "Silent"
s1 = s1.lower().replace(" ", "")
s2 = s2.lower().replace(" ", "")
Counter(s1) == Counter(s2)
```

```python
def is_anagram(s1, s2):
    s1 = s1.lower().replace(" ", "")
    s2 = s2.lower().replace(" ", "")

    if len(s1) != len(s2):
        return False

    freq = {}

    for ch in s1:
        freq[ch] = freq.get(ch, 0) + 1

    for ch in s2:
        if ch not in freq:
            return False
        freq[ch] -= 1
        if freq[ch] < 0:
            return False

    return True
is_anagram(s1,s2)
```

Find the longest word in a sentence.

```python
def longest_word(sentence):
    words = sentence.split()

    if not words:
        return None

    longest = words[0]

    for word in words:
        if len(word) > len(longest):
            longest = word

    return longest
longest_word("I love programming in python")
```
