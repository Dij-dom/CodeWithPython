DataFrame Basics

Create a DataFrame from a dictionary.

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "salary": [50000, 60000, 70000]
}
data
```

```python
import pandas as pd

df = pd.DataFrame(data)
df
```

Select rows based on condition.

```python
df[df["age"] > 30]
```

```python
df[(df["age"] > 25) & (df["salary"] > 55000)]
```

Select specific columns.

```python
df[["name", "salary"]]
```


Rename columns.


```python
df = df.rename(columns={"name": "employee_name", "salary": "income"})
df.columns
```

Change data types.

```python
df.dtypes
```

```python
df["income"] = df["income"].astype(float)
df.dtypes
```
