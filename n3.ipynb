Count frequency of elements in a list.


```python
lst = [x for x in range(9, 130, 10)]
lst.extend([9, 39, 39, 79, 79, 79, 129])
# lst
```

```python
lst
```

```python
count_dict = {}

for item in lst:
    count_dict[item] = count_dict.get(item, 0)+1
count_dict
```

```python
dict = {'a': 1, 'b': 1, 'c': 2}
inverted_dict = {}

for key, value in dict.items():
    inverted_dict[value] = key
inverted_dict
```

Here, a is silently lost, so check for a whether it is present in inverted.

```python
# we create a list of values by default and always keep appending.
inverted_dict = {}
for key, value in dict.items():
    if value not in inverted_dict:
        inverted_dict[value] = []
    inverted_dict[value].append(key)
inverted_dict
```


Find keys with maximum value in a dictionary.



```python
d = {'a': 10, 'b': 20, 'c': 20, 'd': 5}
max_value = None
max_keys = []
for k, v in d.items():
    if max_value is None or v>max_value:
        max_value = v
        max_keys = [k] 
    elif v == max_value:
        max_keys.append(k)
max_keys, max_value
```

Merge two dictionaries and sum values for common keys.


```python
d1 = {'a': 10, 'b': 20, 'c': 30}
d2 = {'b': 5,  'c': 15, 'd': 40}
final_dict = {}

for k1, v1 in d1.items():
    final_dict[k1] = v1

for k2, v2 in d2.items():
    if k2 in final_dict:
        final_dict[k2] +=v2 
    else:
        final_dict[k2] = v2

final_dict
```
